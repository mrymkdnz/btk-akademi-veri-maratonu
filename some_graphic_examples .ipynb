```python
import matplotlib.pyplot as plt
plt.plot([1,2,3,4],[1,4,9,16], 'r+')#önce yatay eksen değerleri, sonra düşey eksen değerleri
plt.xlabel('tamsayılar')
plt.xlabel('tamsayıların karesi')
plt.title('tamsayıların kareleri')
plt.show()
```

```python
import numpy as np
t = np.arange(0.0,5.0,0.2)

plt.plot(t,t,'r--',t,t**2,'bs',t,t**3,'g^')
plt.show()
```

```python
veri = {'a': np.arange(50),
       'c' : np.random.randint(0,50,50),
       'd': np.random.randn(50)}
veri['b'] = veri['a'] + (10* np.random.randn(50))
veri['d'] = np.abs(veri['d']) * 100

plt.scatter('a','b', c='c', s='d', data=veri)
plt.xlabel('a sütunundaki veriler')
plt.ylabel('b sütunundaki veriler')
plt.show()
```

```python
isimler = ['Ahmet','Mehmet', 'Ayşe', 'Fatma']
degerler =[5,25,50,100]
plt.figure(figsize=(9,3))

plt.subplot(131)
plt.bar(isimler,degerler)

plt.subplot(132)
plt.scatter(isimler,degerler)

plt.subplot(133)
plt.plot(isimler,degerler)

plt.suptitle('Bir çok farklı grafik')
plt.show()
```

```python
t = np.arange(0.0,5.0,0.2)
plt.plot(t,t,'r-',linewidth=2.0)
plt.show()
```

```python
t = np.arange(0.0,5.0,0.1)
plt.plot(t,(np.exp(-t)*np.cos(2*np.pi*t)),'k', t,(np.exp(-t)*np.cos(2*np.pi*t)),'bo' )
plt.show()
```

```python
x = np.arange(0,10,0.01)
y=np.exp(x)

plt.figure()

plt.subplot(121)
plt.plot(x,y)
plt.yscale('Linear')
plt.title('Doğrusal')
plt.grid(True)

plt.subplot(122)
plt.plot(x,y)
plt.yscale('Log')
plt.title('Logaritmik')
plt.grid(True)

plt.show()
```
