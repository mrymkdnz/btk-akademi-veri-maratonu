```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

import warnings
warnings.filterwarnings("ignore")
```

```python
veri = pd.read_csv("olimpiyatlar.csv")
veri.head()
```

```python
def sacilimGrafikCiz():
    sns.scatterplot(x='boy',y='kilo',data=veri)
    plt.title('Boy ve Ağırlık İlişkisi - Beyaz Isgara Tema')
    plt.show()
sacilimGrafikCiz()
```

```python
sns.set_style("whitegrid")
sacilimGrafikCiz()
```

```python
sns.scatterplot(x='boy', y='kilo', hue='cinsiyet',style='madalya',size='yas',sizes=(15,200), data=veri),
plt.title('Madalyaya göre boy ve ağırlık dağılımı')
plt.show()
```

```python
sns.displot(veri, x='kilo')
plt.ylabel("Frekans")
plt.title("Ağırlık Histogramı")
plt.show()
```

```python
sns.displot(veri, x="kilo", hue="cinsiyet")
plt.ylabel("Frekans")
plt.title("Cinsiyete göre ayrıştırılmış")
plt.show()
```

```python
sns.displot(veri, x="kilo", col="cinsiyet",multiple="dodge")
plt.show()
```

```python
sns.displot(veri, x="kilo", y="boy", kind="kde")
plt.xlabel("ağırlık")
plt.ylabel("boy")
plt.title("Ağırlık-Boy Histogramı")
plt.show()
```

```python
sns.displot(veri, x="kilo", y="boy",hue="cinsiyet", kind="kde")
plt.xlabel("ağırlık")
plt.ylabel("boy")
plt.title("Cinsiyete Göre Ayrıştırılmış Ağırlık-Boy Histogramı")
plt.show()
```

```python
sns.barplot(x="madalya", y="boy", data=veri)
plt.title("Madalyalara Göre Boy Grafikleri")
plt.show()
```

```python
sns.barplot(x="madalya", y="boy", data=veri, hue="cinsiyet")
plt.title("Madalyalara Göre Boy Grafikleri")
plt.show()
```

```python
sns.catplot(x="madalya", y="yas", hue="cinsiyet", col="sezon", data=veri, kind="bar", height=4, aspect=0.7)
plt.show()
```

```python
sns.boxplot(x="sezon", y="boy", data=veri)
plt.show()
```

```python
sns.boxplot(x="sezon", y="boy",hue="cinsiyet",palette="Set2", data=veri)
plt.show()
```

```python
veri_gecici = veri.loc[:,["yas", "kilo", "boy"]]
sns.boxplot(data=veri_gecici, orient="h", palette="Set2")
plt.show()
```

```python
sns.catplot(x="sezon", y="boy", hue="cinsiyet", col="madalya", data=veri, kind="box", height=4, aspect=0.7)
plt.show()
```

```python
sns.heatmap(veri.corr(),annot=True, linewidths=0.5, fmt='.3f')
plt.show()
```

```python
sns.heatmap(veri.corr(),annot=False, linewidths=0.5, fmt='.3f')
plt.show()
```

```python
sns.violinplot(x="sezon", y="boy", data=veri)
plt.show()
```

```python
sns.violinplot(x="sezon", y="boy",hue="cinsiyet", data=veri)
plt.show()
```

```python
sns.catplot(x="sezon", y="boy",hue="cinsiyet",col="madalya", data=veri, kind="violin",split=True, height=4,aspect=0.7)
plt.show()
```

```python
sns.jointplot(data=veri, x="kilo", y="boy", hue="sezon",kind="kde")
plt.show()
```

```python
g = sns.JointGrid(data=veri,x="kilo", y="boy")
g.plot_joint(sns.histplot)
g.plot_marginals(sns.boxplot)
plt.show()
```

```python
sns.pairplot(veri)
plt.show()
```

```python
g=sns.PairGrid(veri)
g.map_upper(sns.histplot)
g.map_lower(sns.kdeplot,fill=True)
g.map_diag(sns.histplot, kde=True)
plt.show()
```

```python
sns.countplot(x="sehir", data=veri)
plt.xticks(rotation = 90)
plt.show()
```
